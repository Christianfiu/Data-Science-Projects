```python
import pandas as pd
```

```python
url = 'https://raw.githubusercontent.com/dataquestio/project-walkthroughs/master/football_matches/matches.csv'
matches = pd.read_csv(url, index_col = 0)
```

```python
matches['date'] = pd.to_datetime(matches['date'])
```

```python
matches.dtypes
```

```python
#Designates home games to 1 and away games to 2
matches['venue_code'] = matches['venue'].astype('category').cat.codes
```

```python
#creates a numerical value for each team
matches['opp_code'] = matches['opponent'].astype('category').cat.codes
```

```python
#remove colon and minutes from time
#.+ --> '.' means any following character besides newline, '+' means one or more occurences
matches['hour'] = matches['time'].str.replace(':.+', "", regex = True).astype('int')
```

```python
matches['day_code'] = matches['date'].dt.dayofweek
```

```python
matches['target'] = (matches['result'] == 'W').astype('int')
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rf = RandomForestClassifier(n_estimators = 50, min_samples_split=10, random_state =1 )
```

```python
train = matches[matches["date"] < '2022-01-01']
```

```python
test = matches[matches["date"] > '2022-01-01']
```

```python
predictors = ['venue_code', 'opp_code', 'hour', 'day_code']
```

```python
rf.fit(train[predictors], train['target'])
```

```python
preds = rf.predict(test[predictors])
```

```python
from sklearn.metrics import accuracy_score
```

```python
#accuracy of model
acc = accuracy_score(test['target'], preds)
```

```python
#create dataframe
combined = pd.DataFrame(dict(actual = test['target'], prediction = preds), index = test.index)
```

```python
group_matches = matches.groupby('team')
```

```python
group = group_matches.get_group('Manchester City')
```

```python
def rolling_averages(group, cols, new_cols):
  group = group.sort_values('date')
  #closed = left, pandas will compute rolling averages will use knowledge of the future if this is not set
  #if we're on week 4, we only want the info from weeks 1,2,3
  rolling_stats = group[cols].rolling(3, closed = 'left').mean()
  group[new_cols] = rolling_stats
  #dropna will remove missing value rows
  group = group.dropna(subset = new_cols)
  return group
```

```python
cols = ['gf', 'ga', 'sh', 'sot', 'dist', 'fk', 'pk', 'pkatt']
new_cols = [f'{c}rolling' for c in cols]
```

```python
rolling_averages(group, cols, new_cols)
```

```python
#groups by team
#
matches_rolling = matches.groupby('team').apply(lambda x: rolling_averages(x, cols, new_cols))
```

```python
matches_rolling
```

```python
matches_rolling = matches_rolling.droplevel('team')
```

```python
matches_rolling
```

```python
#assigns indiv values to rows
matches_rolling.index = range(matches_rolling.shape[0])
```

```python
matches_rolling
```

```python
def make_predictions(data, predictors):
  train = data[data["date"] < '2022-01-01']
  test = data[data["date"] > '2022-01-01']
  rf.fit(train[predictors], train['target'])
  preds = rf.predict(test[predictors])
  combined = pd.DataFrame(dict(actual = test['target'], prediction = preds), index = test.index)
  precision = precision_score(test['target'], preds)
  return combined, precision
```

```python
combined, precision = make_predictions(matches_rolling, predictors + new_cols)
```

```python
#combining
class MissingDict(dict):
  __missing__ = lambda self, key : key

map_values = {
    'Brighton and Hove Albion': 'Brighton',
    'Manchester United' : 'Manchester Utd',
    'Newcastle United' : 'Newcastle Utd',
    'Tottenham Hotspur' : 'Tottenham',
    'West Ham United' : 'West Ham',
    'Wolverhampton Wanderers' : 'Wolves'
}

mapping = MissingDict(**map_values)
```

```python
combined['new_team'] = combined['team'].map(mapping)
combined
```
